```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

## Scatter plot

```python
# Scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
#  relplot -> figure level function.
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
# additional parameters
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

```python
# Facet plot -> multiple graphs inside figure using 'col','row'
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex', row='time')
```

```python
gap = px.data.gapminder()
gap
```

```python
temp = gap[gap['country']=='India']
tempdf = gap[gap['country'].isin(['India', 'Germany', 'Zimbabwe'])]
```

```python
temp
```

```python
# Axes level
sns.lineplot(data=temp, x='year', y='lifeExp')
```

```python
# figure level
sns.relplot(data=temp, x='year', y='lifeExp', kind='line')
```

```python
tempdf
```

```python
# multi line graph
sns.relplot(data=tempdf, x='year', y='lifeExp', kind='line', hue='country', style='continent')
```

```python
# Set the style and context of seaborn
sns.set(style="whitegrid")

# Increase the scale of the plot
sns.set(rc={'figure.figsize':(10,6)})

# Create the boxplot with more points on the y-axis
sns.boxplot(data=tips, y='total_bill')
```

```python
sns.boxplot(data=tips, y='total_bill')

# Set the ticks on the y-axis
plt.yticks(range(5, 26, 5))  # This will set ticks from 5 to 25 in steps of 5

# Show the plot
plt.show()
```

```python
sns.boxplot(data=tips, y='total_bill')

# Get the current Axes object
ax = plt.gca()

# Set the ticks on the y-axis
ax.set_yticks(range(5, 26, 5))  # This will set ticks from 5 to 25 in steps of 5

# Show the plot
plt.show()
```

```python
ax = sns.catplot(data=tips, x='total_bill', kind='box')

# Set the ticks on the y-axis
ax.set_xticks(range(1, 56, 1))  # This will set ticks from 5 to 55 in steps of 5

# Show the plot
ax
```
